```python
# Importing the Dataset Sabli Gate Hapur (2)
data_02 = pd.read_csv("C:/Users/ritam/OneDrive/Desktop/assignmentsforbusinessanalyst (3)/Sabli Gate Hapur (2).csv")
data_02
```

```python
num_empty_columns = data_02.isna().all().sum() # Count the number of columns with all missing values


print("Number of columns with absolutely no data:", num_empty_columns)
```

```python
data_02 = data_02.dropna(axis=1, how='all') # Drop columns with all missing values in the original DataFrame

data_02.shape
```

```python
data_02 = data_02.dropna(subset=['Order_Source'])

num_unique_order_sources = data_02['Order_Source'].nunique() # Counting the number of unique values in the 'Order_Source' column

order_source_counts = data_02['Order_Source'].value_counts() # Get the counts of each unique value in the 'Order_Source' column

print("Number of unique values in 'Order_Source':", num_unique_order_sources)
print("Unique values in 'Order_Source' and their counts:")
print(order_source_counts)
```

```python
# Converting 'Business_Date' column to datetime format
data_02['Business_Date'] = pd.to_datetime(data_02['Business_Date'], format='%d-%m-%Y')

# detecting which months data available in the dataset
months_present = data_02['Business_Date'].dt.month.dropna().unique() #filtering out missing values

month_names = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June', 
               7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}

# Converting month numbers to month names
months_present_names = [month_names[month] for month in months_present]

# Print the months present in the data
print("Months present in the data are:", months_present_names)
```

```python
# Calculating the 'Food_Cost': 40% of revenue - creating a column for Food cost
data_02['Food_Cost'] = data_02['Net_Amount'] * 0.4
```

```python
# Create a dictionary to map order sources to commission rates
commission_rates = {
    'POS': 0.23,              # for POS commission 23% 
    'zomato': 0.237,          # for zomato commission 23.7%
    'swiggy': 0.223,          # for swiggy commission 22.3%
    'dotpe': 0.23            # assuming dotpe commision is also 23%, because it is not mentioned
}

# Function to calculate commission based on order source
def calculate_commission(row):
    return row['Net_Amount'] * commission_rates[row['Order_Source']]

# Apply the function to create the new column
data_02['Commission_on_delivery_sales'] = data_02.apply(calculate_commission, axis=1)
```

```python
# Calculating the 'Total_Cost' column
data_02['Total_Cost'] = (data_02['Materials_Cost'] +
                                  data_02['Supplies_Cost'] +
                                  data_02['Food_Cost'] +
                                  data_02['Commission_on_delivery_sales'])
```

```python
# Filtering data for September
data_september_02 = data_02[data_02['Business_Date'].dt.month == 9]     # Calculating total revenue for September
total_revenue_september_02 = data_september_02['Net_Amount'].sum()     # Calculating total cost for September
total_cost_september_02 = data_september_02['Total_Cost'].sum()     # Calculating total profit for September
total_profit_september_02 = total_revenue_september_02 - total_cost_september_02

# Filtering data for October
data_october_02 = data_02[data_02['Business_Date'].dt.month == 10]     # Calculating total revenue for October
total_revenue_october_02 = data_october_02['Net_Amount'].sum()     # Calculating total cost for October
total_cost_october_02 = data_october_02['Total_Cost'].sum()     # Calculating total profit for October
total_profit_october_02 = total_revenue_october_02 - total_cost_october_02

# Filtering data for November
data_november_02 = data_02[data_02['Business_Date'].dt.month == 11]     # Calculating total revenue for October
total_revenue_november_02 = data_november_02['Net_Amount'].sum()     # Calculating total cost for October
total_cost_november_02 = data_november_02['Total_Cost'].sum()     # Calculating total profit for October
total_profit_november_02 = total_revenue_november_02 - total_cost_november_02

# Printing the results
print("Total revenue for September:", total_revenue_september_02)
print("Total cost for September:", total_cost_september_02)
print("Total profit for September:", total_profit_september_02)

print("\nTotal revenue for October:", total_revenue_october_02)
print("Total cost for October:", total_cost_october_02)
print("Total profit for October:", total_profit_october_02)

print("\nTotal revenue for November:", total_revenue_november_02)
print("Total cost for November:", total_cost_november_02)
print("Total profit for November:", total_profit_november_02)
```

```python
import pandas as pd

# Creating a dictionary to hold the data
data_dict = {
    'Month': ['September', 'October', 'November'],
    'Revenue': [total_revenue_september_02, total_revenue_october_02, total_revenue_november_02],
    'Total Cost': [total_cost_september_02, total_cost_october_02, total_cost_november_02],
    'Profit': [total_profit_september_02, total_profit_october_02, total_profit_november_02]
}

# Creating a DataFrame from the dictionary
monthly_summary = pd.DataFrame(data_dict)

# Printing the DataFrame
print(monthly_summary)
```

```python
#Analysis based on Order source for Buger Singh Outlet
# Define a function to calculate total revenue, total cost, and total profit for each order source and month
def calculate_metrics(data, order_source):
    data_filtered = data[data['Order_Source'] == order_source]
    total_revenue = data_filtered['Net_Amount'].sum()
    total_cost = data_filtered['Total_Cost'].sum()
    total_profit = total_revenue - total_cost
    return total_revenue, total_cost, total_profit

# Create a list to store the results for each order source and month
results = []

# Loop through each order source
for order_source in ['POS', 'zomato', 'swiggy', 'dotpe']:
    # Calculate metrics for September
    revenue_september, cost_september, profit_september = calculate_metrics(data_september_02, order_source)
    
    # Calculate metrics for October
    revenue_october, cost_october, profit_october = calculate_metrics(data_october_02, order_source)
    
    # Calculate metrics for November
    revenue_november, cost_november, profit_november = calculate_metrics(data_november_02, order_source)
    
    # Append the results to the list
    results.append({
        'Order_Source': order_source,
        'September_Revenue': revenue_september,
        'September_Cost': cost_september,
        'September_Profit': profit_september,
        'October_Revenue': revenue_october,
        'October_Cost': cost_october,
        'October_Profit': profit_october,
        'November_Revenue': revenue_november,
        'November_Cost': cost_november,
        'November_Profit': profit_november
    })

# Create a DataFrame from the results
result_analysis = pd.DataFrame(results)
```
